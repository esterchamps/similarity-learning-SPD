```python
import numpy as np
from scipy import ndimage
from scipy.linalg import expm
import os
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def load_image(image_path):
    image = cv.imread(image_path)
    gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    image_array = np.array(gray_image)

    return image_array
```

```python
img = load_image('C:\\Users\\ester\\faculdade\\tcc\\database_gpds\\SignatureGPDSSyntheticSignaturesManuscripts\\firmasSINTESISmanuscritas\\002\\cf-002-13.jpg')
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
thr, otsu_result = cv.threshold(img, 0, 1, cv.THRESH_BINARY + cv.THRESH_OTSU)
otsu_inverted = otsu_result^1
```

```python
plt.imshow(otsu_inverted, cmap='gray', vmin=0, vmax=1)
```

```python
#img_thinn = ndimage.binary_opening(otsu_inverted, iterations=3)
```

```python
img_thinn = ndimage.binary_erosion(otsu_inverted, iterations=1)
```

```python
plt.imshow(img_thinn, cmap='gray', vmin=0, vmax=1)
```

```python
res = img * img_thinn
```
